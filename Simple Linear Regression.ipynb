```python
w = 0
b = 0

def plot_pred(w,b):
    y_pred = w*x + b
    plt.plot(x,y_pred,color="blue",label="predict line")
    plt.scatter(x,y,marker="x",color="red",label="actual data")
    plt.title("Square Feet vs Price")
    plt.xlabel("Square Feet")
    plt.ylabel("Price (k)")
    plt.xlim([0,200])
    plt.ylim([0,1000])
    plt.legend()
    plt.show()

plot_pred(6,0)
```

```python
#cost function
w = 6
b = 0

y_pred = w*x + b
cost = (y - y_pred)**2
cost.sum()/len(x)
# cost.mean()
```

```python
def compute_cost(x,y,w,b):
    y_pred = w*x + b
    cost = (y - y_pred)**2
    cost = cost.sum()/len(x)

    return cost
```

```python
compute_cost(x,y,6,0)
```

```python
#b=0 w=100~100 cost=?

costs = []
for w in range(-100,101):
    cost = compute_cost(x,y,w,0)
    costs.append(cost)

costs
```

```python
plt.plot(range(-100,101),costs)
plt.title("cost function w=-100 to 101 b=0")
plt.xlabel("w")
plt.ylabel("cost")
plt.show()
```

```python
import numpy as np
```

```python
ws = np.arange(-100,101)
bs = np.arange(-100,101)
costs = np.zeros((201,201))

i = 0
for w in ws:
    j = 0
    for b in bs:
        cost = compute_cost(x,y,w,b)
        costs[i,j] = cost
        j = j + 1
    i = i + 1

costs
```

```python
ax = plt.axes(projection = "3d")
ax.view_init(45,-120)
b_grid, w_grid = np.meshgrid(bs,ws)
ax.plot_surface(w_grid,b_grid,costs,cmap="Spectral_r",alpha=0.7)
ax.plot_wireframe(w_grid,b_grid,costs,color="black",alpha=0.1)

ax.set_title("w b co-responding cost")
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("costs")

w_index, b_index = np.where(costs == np.min(costs))
ax.scatter(ws[w_index],bs[b_index],costs[w_index,b_index],color="red",s=40)

plt.show()
print(f"when w={ws[w_index]},b={bs[b_index]} will appear min cost:{costs[w_index,b_index]}")
```
